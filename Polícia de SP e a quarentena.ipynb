### 2.1  2020

Primeiramente é necessário escolher o site do qual será extraído as informações, neste caso escolhemos o site do governo de São Paulo que possuiu os dados necessários.
É necessário verificar se a página está em HTML e não em Javascript. Para saber se o conteúdo está em HTML ou em Javascript devemos fazer a inspeção (clicando com o botão direito) na página e observar as etiquetas. Em seguida, temos que verificar se existe código HTML entre as etiquetas . Sabemos que se trata de Javascript quando dentro das etiquetas aparecem somente etiquetas como "script>nome-de-um-arquivo.js</script"



```python
url_pagina_html_2020="https://www.ssp.sp.gov.br/Estatistica/Pesquisa.aspx"
```

Com a biblioteca requests fazemos o download da página

```python
pagina_html_2020 = requests.get(url_pagina_html_2020)
pagina_html_2020
```

Com isso criamos um objeto soup que mostra o conteúdo da página html formatado para uma melhor visualização.

```python
soup = BeautifulSoup(pagina_html_2020.content, 'html.parser')
#soup.prettify()
```

Após isto, selecionamos a parte que nos interessa do código.

```python
lista_de_tabelas_html_2020 = soup.find_all(id="conteudo_repAnos_gridDados_0")
```

```python
tb_casos_por_mes_tag_2020 = lista_de_tabelas_html_2020[0]
```

É importante verificar o tipo da variável, precisa ser do tipo <bs4.element.Tag>

```python
type(tb_casos_por_mes_tag_2020)
```

```python
linhas_tb_2020 = tb_casos_por_mes_tag_2020.find_all('tr')
```

```python
lin_2020 = linhas_tb_2020[1].get_text("\n")
lin_2020
```

```python
lin_2020 = lin_2020.strip("\n").strip(" ")
lin_2020
```

```python
lin_2020=lin_2020.replace("\n","\t")
lin_2020
```

```python
lin_2020=lin_2020.replace(".","")
lin_2020
```

```python
lin_2020 = lin_2020.rstrip("\t")
lin_2020
```

```python
def prep_linha_2020(linha_2020):
    # Remove os caracteres \n no começo e final da string
    # Em seguida, remove os espaços
    linha_2020 = linha_2020.get_text("\n").strip("\n").strip(" ")
    
    linha_2020 = linha_2020.replace(".","")

    
    # Substitui os caracteres \n no meio da string por \t
    linha_2020 = linha_2020.replace("\n","\t")

    
    # Remove o \t ao final da linha
    linha_2020 = linha_2020.rstrip("\t")
    

    
    return linha_2020
```

```python
ini_tb = 1
fim_tb = 13
```

```python
texto_final_2020 = "Natureza\tJaneiro_2020\tFevereiro_2020\tMarço_2020\tAbril_2020\tMaio_2020\tJunho_2020\tJulho_2020\tAgosto_2020\tStembro_2020\tOutubro_2020\tNovembro_2020\tDezembro_2020\tTotal_2020\n"
for i in range(ini_tb, fim_tb+1):
    linha = prep_linha_2020(linhas_tb_2020[i])
    texto_final_2020 += linha
    texto_final_2020 += "\n"
print(texto_final_2020)
```
